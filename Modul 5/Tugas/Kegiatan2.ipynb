```python
import matplotlib.pyplot as plt
import numpy as np

data_transaksi = [
    ("Produk A", 50, 10),
    ("Produk B", 30, 25),
    ("Produk C", 20, 30),
    ("Produk D", 60, 8),
    ("Produk E", 40, 15),
    ("Produk F", 70, 5),
]

# TODO 1: Ekstrak harga produk dan jumlah produk terjual untuk visualisasi pertama
harga_produk = list(map(lambda x: x[1], data_transaksi))
jumlah_terjual = list(map(lambda x: x[2], data_transaksi))

# TODO 2: Buat scatter plot untuk hubungan antara harga produk dan jumlah produk terjual
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(harga_produk, jumlah_terjual, color='magenta')
plt.xlabel('Harga Produk')
plt.ylabel('Jumlah Produk Terjual')
plt.title('Scatter Plot Hubungan Harga Produk dan Jumlah Produk Terjual')

# TODO 3: Hitung total pendapatan untuk setiap produk
pendapatan_produk = list(map(lambda x: x[1] * x[2], data_transaksi))

# TODO 4: Tambahkan bar chart untuk menyajikan pendapatan produk
plt.subplot(1, 2, 2)
bar_width = 0.4
index = np.arange(len(data_transaksi))
bars = plt.bar(index, pendapatan_produk, bar_width, label='Pendapatan', color='pink')
plt.xlabel('Produk')
plt.ylabel('Pendapatan')
plt.title('Bar Chart Pendapatan Produk')
plt.xticks(index, [produk[0] for produk in data_transaksi])
plt.legend()

plt.tight_layout()
plt.show()
```

# Kesimpulan Hasil Plot Data

1. Scatter Plot:
    * Terdapat variasi dalam hubungan antara harga produk dan jumlah produk terjual.
    * Produk D, yang memiliki harga tinggi, memiliki jumlah produk terjual yang relatif rendah.
    * Produk F, yang memiliki harga tinggi, memiliki jumlah produk terjual yang sangat rendah.

2. Bar Chart:
    * Produk F memiliki pendapatan yang tinggi meskipun jumlah produk terjualnya rendah, karena harganya yang tinggi.
    * Produk C memiliki jumlah produk terjual yang tinggi, tetapi pendapatannya tidak setinggi Produk F karena harganya yang lebih rendah.
    
3. Pendapatan vs. Jumlah Produk Terjual:
    * Meskipun Produk C memiliki jumlah produk terjual tertinggi, pendapatannya tidak mendominasi karena harga produk yang lebih rendah. 
    * Produk D memiliki harga produk tinggi, tetapi pendapatan rendah karena jumlah produk terjual yang sedikit.

4. Variasi Produk:
    * Terdapat variasi yang signifikan dalam harga, jumlah produk terjual, dan pendapatan antar produk.
