# Percobaan 1

Gambar garis dalam diagram dari posisi (0,0) ke posisi (6,25):

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 25])

plt.plot(xpoints, ypoints)
plt.show()
```

## Modifikasi

Gambar garis dalam diagram dari posisi (1, 3) ke posisi (8, 10):

```python
xpoints_new = np.array([1, 8])
ypoints_new = np.array([3, 10])

plt.plot(xpoints_new, ypoints_new)
plt.show()
```
