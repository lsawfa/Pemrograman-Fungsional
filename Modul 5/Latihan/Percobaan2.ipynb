```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoints = np.array([1, 8, 10])
ypoints = np.array([3, 10, 5])

plt.figure(figsize=(5,5))
plt.plot(xpoints, ypoints, color='red')
plt.xlim([0,15])
plt.ylim([0,15])
plt.show()
```

## Modifikasi

```python
fig = plt.figure(figsize=(5, 5))

plt.plot(xpoints, ypoints, 
         linewidth=2,    
         color='green',  
         linestyle='--', 
         marker='o',     
         markersize=8,   
         markerfacecolor='red',  
         markeredgecolor='black',
         label='GMV Data'
         )

plt.title('Monthly GMV Year 2023', loc='center', pad=20, fontsize=15, color='black')
plt.xlabel('Month', loc='right', fontsize=12)
plt.ylabel('Order Month', loc='top', fontsize=12)

plt.xlim([0, 13])
plt.ylim([0, 15])

plt.grid(color='yellow', linestyle=':', linewidth=0.5)

plt.text(0.17, 0.72, 'Temporary GMV data is only up to the 10th month', transform=fig.transFigure, color='yellow')

plt.legend()

plt.show()
```
