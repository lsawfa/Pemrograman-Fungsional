## Scatter Plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 7, 2, 8, 5]

plt.scatter(x, y)
plt.show()
```

```python
plt.scatter(x, y, 
            color='yellow',    
            marker='^',        
            s=100,             
            edgecolors='black', 
            alpha=0.8     
            )

plt.title('Scatter Plot dengan Variasi')
plt.xlabel('Nilai x')
plt.ylabel('Nilai y')

plt.show()
```
