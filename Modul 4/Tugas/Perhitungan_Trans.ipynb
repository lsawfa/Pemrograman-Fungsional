## Persamaan Garis Sebelum Transformasi

\[ y = mx + c \]

   \[ 8 = (1) (1) + c \]
   
   \[ 8 = 1 + c \]

   \[ c = 8 - 1\]

   \[ c = 7 \]

   \[ y = x + 7 \]

   Jadi persamaan garis setelah transformasi adalah
   $$
   y = x + 7
   $$
